# Google mobility trends
We analyse google mobility data during COVID19. 
The main idea is to show how mobility has changed during COVID19.
We use the open data from https://github.com/datasciencecampus/google-mobility-reports-data 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
filepath_loc ="C:/Users/lyubo/Documents/DATA_networks/covid19/google-mobility-reports-data-master/csvs/international_local_area_trends_UK_DE_FR_ES_IT_US_SE.csv"
filepath_nat ="C:/Users/lyubo/Documents/DATA_networks/covid19/google-mobility-reports-data-master/csvs/international_national_trends_UK_DE_FR__ES_IT_US_SE.csv"

data_mob_loc = pd.read_csv(filepath_loc)
print(data_mob_loc.shape)
data_mob_loc.head()
```

```python
import numpy as np
# find unique values of Country 
un_count = np.unique(data_mob_loc['Country'].values)

print(un_count)
```

```python
# get only df for France
df_France = data_mob_loc[data_mob_loc['Country']=="France"]
df_France.head()
```

```python
import numpy as np
# find unique values of Country 
un_count = np.unique(df_France['location'].values)

print('unique locations for France',un_count)
```

```python
# create Ile de France mobility data 

df_ile_de_fr = df_France[df_France['location'] =='ÃŽle-de-France']

print(df_ile_de_fr.shape)
df_ile_de_fr.head(6)

```

# Plotting mobility trends

```python

plt.figure(figsize=(20,10))

for ind in range(0,6): # plot each category
    row = df_ile_de_fr.iloc[ind,4:48].values
    #print(row)
    plt.plot(row, label = df_ile_de_fr['category'].iloc[ind])

    plt.xlabel("time")
    plt.ylabel("value")
plt.legend(title ='mobility trend' )
plt.title("mobility trend in Ile de France, 2020-02-16 - 2020-03-29")
plt.show()

```

# Plotting in national trends


```python
filepath_nat ="C:/Users/lyubo/Documents/DATA_networks/covid19/google-mobility-reports-data-master/csvs/international_national_trends_UK_DE_FR__ES_IT_US_SE.csv"

data_mob_nat = pd.read_csv(filepath_nat)
print('shape', data_mob_nat.shape)
data_mob_nat.head()
import numpy as np
# find unique values of Country 
un_count = np.unique(data_mob_nat['Country'].values)

print(un_count)


# get only df for France
df_France_nation = data_mob_nat[data_mob_nat['Country']=="France"]
df_France_nation.head()
```

```python

plt.figure(figsize=(20,10))

for ind in range(0,6): # plot each county data
    row = df_France_nation.iloc[ind,3:46].values
    #print(row)
    plt.plot(row, marker ='o', label = df_France_nation['Category'].iloc[ind])

    plt.xlabel("time")
    plt.ylabel("value")
plt.legend(title ='Mobility index')
plt.title("Mobility trend in France")
plt.show()

```
